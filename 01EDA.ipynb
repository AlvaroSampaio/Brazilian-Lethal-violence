## 1- Valores nulos
Vamos começar da coluna `where_description`. Ela possui 11 valores nulos, no entanto eles podem ser facilmente preenchidos com uma pesquisa, porque as informações de latitude e longitude estão disponíveis no dataframe

```python
df[df['where_description'].isnull()][['geom_wkt']]
```

```python
# Definindo as condições e os valores a serem preenchidos
conditions_values = {
    'POINT (-43.2075 -22.90278)': 'Rio de Janeiro town',
    'POINT (-38.9 -3.6)': 'Fortaleza',
    'POINT (-38.54306 -3.71722)': 'Fortaleza',
    'POINT (-40 -5)': 'Nossa Senhora do Livramento',
    'POINT (-40.21222 -3.46056)': 'Santana do Acaraú',
    'POINT (-60.01282 -3.04361)': 'Manaus'
}
# Preenchendo os valores nulos na coluna 'where_description' com base nas condições
for geom_wkt, description in conditions_values.items():
    df.loc[(df['geom_wkt'] == geom_wkt) & (df['where_description'].isnull()), 'where_description'] = description
```

A coluna souce_date em algumas linhas está devolvendo dois valores de datas, mas não é possível que isso aconteça. Vamos corrigir está inconsistência

Aqui estamos definindo uma função que fará a varredura da coluna para detectar o tipo de inconsistência descrita e aplicar a correção

```python
def convert_to_date(date_str: str) -> Union[pd.Timestamp, None]:
    """
    Convert a string to a datetime object. Tries different formats for conversion.

    Args:
        date_str (str): The input date string.

    Returns:
        Union[pd.Timestamp, None]: The converted datetime object or None if conversion fails.
    """
    try:
        return pd.to_datetime(date_str, format='%Y%m%d')
    except ValueError:
        try:
            return pd.to_datetime(date_str)
        except ValueError:
            return None
def preprocess_date_column(df: pd.DataFrame, column: str) -> pd.DataFrame:
    """
    Preprocess the specified date column in the dataframe by performing the following steps:
    1. Convert all values to strings.
    2. Split the string by ';' and keep only the first part if there are multiple dates.
    3. Convert the string dates to datetime format.

    Args:
        df (pd.DataFrame): The input dataframe with a date column.
        column (str): The name of the column to be processed.

    Returns:
        pd.DataFrame: The dataframe with the specified column processed and converted to datetime.
    """
    # Convert column values to strings
    df[column] = df[column].astype(str)

    # Keep only the first date if there are multiple dates separated by a semicolon
    df[column] = df[column].apply(lambda x: x.split(';')[0])

    # Apply the conversion to datetime
    df[column] = df[column].apply(convert_to_date)

    return df

# Example usage:
# Assuming 'source_date' is the date column in the DataFrame df
df = preprocess_date_column(df, 'source_date')
```

A coluna souce_article possui algumas linhas com problemas semelhantes. No caso algumas linhas guardam informações duplicadas, como podemos ver na terceira linha do dataframe

```python
df['source_office'].head()
```

Definimos uma função para identificar e corrigir o problema

```python
def remove_duplicates(row: pd.Series) -> str:
    """
    Removes duplicate entries from the 'source_office' field in a row of a DataFrame.
    
    Parameters:
    row (pd.Series): A row from a pandas DataFrame.
    
    Returns:
    str: A string with duplicate entries removed, joined by ';', 
         or the original value if 'source_office' is not a string.
    """
    if isinstance(row['source_office'], str):
        # Split the 'source_office' string by ';' and convert to a set to remove duplicates
        unique_values = set(row['source_office'].split(';'))
        # Join the unique values back into a string separated by ';'
        return ';'.join(unique_values)
    else:
        # If 'source_office' is not a string, return it as is
        return row['source_office']

# Apply the remove_duplicates function to each row in the 'source_office' column of the DataFrame
df['source_office'] = df.apply(remove_duplicates, axis=1)
```
